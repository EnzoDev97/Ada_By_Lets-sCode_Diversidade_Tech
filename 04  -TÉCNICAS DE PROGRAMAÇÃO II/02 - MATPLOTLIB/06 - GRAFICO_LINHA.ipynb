```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [n ** 2 for n in x]

plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [n ** 2 for n in x]
y2 = [n ** 3 for n in x]

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
#Podemos acessar cada eixo usando a notação de colchetes, e invocar a função de plot normalmente. 
#Exemplificamos a chamada dos eixos dentro da figura conforme abaixo

fig, axs = plt.subplots(2, 2, figsize=(12,5))

axs[0,0].text(0.4, 0.4, 'Gráfico[0,0]')
axs[0,1].text(0.4, 0.4, 'Gráfico[0,1]')
axs[1,0].text(0.4, 0.4, 'Gráfico[1,0]')
axs[1,1].text(0.4, 0.4, 'Gráfico[1,1]')

plt.show()
```

```python
# No exemplo utilizado anteriormente podemos plotar da seguinte forma em vários eixos.



x = [1, 2, 3, 4, 5]
y1 = [n ** 2 for n in x]
y2 = [n ** 3 for n in x]

fig, axes = plt.subplots(nrows=1, ncols=2) # Configura a figura e os eixos de criação dos gráficos
axes[0].plot(x, y1)
axes[1].plot(x, y2)

plt.show()

# Note que ao passar nrows=1, espertamente, o matplotlib não devolve uma matriz e sim uma lista, por isso não se faz necessário passar dois valores de posição, apenas um.

```

```python
#O parâmetro ls (linestyle) altera o estilo da linha do gráfico.

x = [1, 2, 3, 4, 5]
y1 = [n ** 2 for n in x]
y2 = [n ** 3 for n in x]

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1, ls='--', c='red')
axes[1].plot(x, y2, ls=':', c='#FF00FF')

plt.show()
```
