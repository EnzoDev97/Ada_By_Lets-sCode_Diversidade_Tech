```python
# 02 -Os parâmetros xlabel e ylabel alteram, respectivamente, os nomes dados aos eixos x e y. 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [n ** 2 for n in x]

plt.plot(x, y)

plt.xlabel('x') # Altera o nome do eixo x
plt.ylabel('f(x)') # Altera o nome do eixo y

plt.show()
```
