# 03 - Utilize a função PALETTE para alterar as cores do gráfico.

_O seaborn possui um recurso que auxilia no design dos gráficos, a paleta de cores. Este recurso mantém a harmonia entre as cores escolhidas para plotagem dos gráficos._

```python
import seaborn as sns 
x = [1, 2, 3, 4, 5 ,6]
y = [10, 5, 3, 4, 6, 8]
z = ['a', 'b', 'c', 'c', 'a', 'a']
sns.scatterplot(x=x, y=y, hue=z, palette='viridis'); # Escolhe a paleta de cores 'viridis'
```
