```python
# importando a biblioteca
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importando a base 'flights' do seaborn
flights = sns.load_dataset('flights')
print(flights.head())
print('_'*100,'\n')

# importando a base 'penguins' do seaborn

penguins = sns.load_dataset('penguins')
penguins.head()
```

# 1 - Qual a relação entre o comprimento e a profundidado do bico?

```python
# visualizando com seaborn - scatterplot

sns.scatterplot(data=penguins,
                x='bill_length_mm',
                y='bill_depth_mm')

plt.show()
```

# 1.1 - segmentando por sexo

```python
# segmentando por sexo

sns.scatterplot(data=penguins,
                x='bill_length_mm',
                y='bill_depth_mm',
                hue='sex')

plt.show()
```

# 1.2 -  segmentando por espécie

```python
# segmentando por espécie

sns.scatterplot(data=penguins,
                x='bill_length_mm',
                y='bill_depth_mm',
                hue='species')

plt.show()
```

# 1.3 - segmentando por espécie e sexo com o parâmetro style

```python
# segmentando por espécie e sexo com o parâmetro style

sns.scatterplot(data=penguins,
                x='bill_length_mm',
                y='bill_depth_mm',
                hue='species',
                style='sex')

plt.show()
```

* visualizando com seaborn - regplot

```python
sns.regplot(data=penguins,
            x='bill_length_mm',
            y='bill_depth_mm')
plt.show()
```

* visualizando com seaborn - lmplot
* segmentando por espécie

```python
# visualizando com seaborn - lmplot
# segmentando por espécie

sns.lmplot(data=penguins,
            x='bill_length_mm',
            y='bill_depth_mm',
            hue='species')
plt.show()
```

# 2 -  Pairplot para Analise Exploratória

## 2.1 - Observando a relação entre todas variáveis numéricas

```python
# Observando a relação entre todas variáveis numéricas

sns.pairplot(penguins)
plt.show()
```

## 2.2 - Observando a relação entre todas variáveis numéricas segmentado por sexo

```python
# Observando a relação entre todas variáveis numéricas segmentado por sexo

sns.pairplot(penguins, hue='sex')
plt.show()
```

## 2.3 - Observando a relação entre todas variáveis numéricas segmentado por espécie

```python
# Observando a relação entre todas variáveis numéricas segmentado por espécie

sns.pairplot(penguins, hue='species')
plt.show()
```
