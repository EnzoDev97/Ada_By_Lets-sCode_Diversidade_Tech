### 01 - COSNTRUINDO PRIMEIRO GRAFÍCO Seaborn

```python
import seaborn as sns 

x = [1, 2, 3, 4, 5 ,6]
y = [10, 5, 3, 4, 6, 8]

sns.scatterplot(x=x, y=y);
```

### 1.1 - Comparando o gráfico acima com o que vimos no matplotlib.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5 ,6]
y = [10, 5, 3, 4, 6, 8]

plt.scatter(x, y);
```

_Você pode notar que o gráfico é extremamente semelhante ao proveniente do matplotlib, o que não é por acaso, visto que o seaborn é baseado no próprio matplotlib._
