### 09 -  Histograma

```python
import seaborn as sns
import numpy as np
import pandas as pd

x = np.random.rand(100) # Cria 100 números aleatórios

sns.histplot(x);
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)

sns.histplot(df['sepal width (cm)']);
```
