### Carregando as Bases de Dados

```python
# importando a biblioteca
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importando a base 'flights' do seaborn
flights = sns.load_dataset('flights')
flights.head()



```

# 1 - Visualizando o total de passageiros por ano com seaborn - lineplot

```python

sns.lineplot(data=flights, x='year', y='passengers', estimator='sum',color ='red')
plt.show()
```

### 2 - Qual a média de passageiros por ano?

```python
# visualizando a média de passageiros por ano com seaborn - lineplot

sns.lineplot(data=flights, x='year', y='passengers', estimator='mean', color='green')
plt.show()
```

### 3 - Qual a quantidade de passageiros por ano, segmentado por mês?

```python
# visualizando com seaborn - lineplot

sns.lineplot(data=flights, x='year', y='passengers', hue='month', estimator='sum')
plt.show()
```
