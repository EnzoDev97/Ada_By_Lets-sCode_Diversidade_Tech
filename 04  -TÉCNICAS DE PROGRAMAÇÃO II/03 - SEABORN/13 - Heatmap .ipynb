# 13 - Heatmap

```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)

sns.heatmap(df.corr());
```

```python
sns.heatmap(df.corr(), cmap='coolwarm', vmin=-1, vmax=1, annot=True);
```
