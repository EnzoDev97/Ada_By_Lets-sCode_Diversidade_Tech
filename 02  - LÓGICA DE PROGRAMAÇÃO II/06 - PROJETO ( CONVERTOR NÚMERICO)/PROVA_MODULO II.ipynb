# DESAFIO DA DISCIPLINA

#### Contexto - Problema: Você foi designado para fazer o tratamento dos dados de um formulário na empresa em que trabalha. O campo salário do formulário, que deveria ser númerico, não está padronizado corretamente, tem numeros escritos por extenso. Seu trabalho é criar uma lib que receba um numero escrito por extenso e converta em valor numérico. 

#### Exemplos: 

##### Entrada: Mil duzentos e cinquenta ---- Saída: 1250

##### Entrada: 1k ---- Saída: 1000

##### Entrada: 1.3k ---- Saída: 1300

```python
# Dicionários para acesso durante a troca de números

dict_numbers = {
    'um': 1,
    'dois': 2,
    'tres': 3,
    'quatro': 4,
    'cinco': 5,
    'seis': 6,
    'sete': 7,
    'oito': 8,
    'nove': 9,
    'dez': 10,
    'onze': 11,
    'doze': 12,
    'treze': 13,
    'quatorze': 14,
    'quinze': 15,
    'dezesseis': 16,
    'dezessete': 17,
    'dezoito': 18,
    'dezenove': 19,
    'vinte': 20,
    'trinta': 30,
    'quarenta': 40,
    'cinquenta': 50,
    'sessenta': 60,
    'setenta': 70,
    'oitenta': 80,
    'noventa': 90,
    'cem': 100,
    'cento': 100,
    'duzentos': 200,
    'trezentos': 300,
    'quatrocentos': 400,
    'quinhentos': 500,
    'seiscentos': 600,
    'setecentos': 700,
    'oitocentos': 800,
    'novecentos': 900,
}

dict_unidades = { 'mil': 1000, 'milhao': 1000000,  'milhoes': 1000000,  'bilhao': 1000000000, 'bilhoes': 1000000000 }

dict_str = { 'k': 1000, 'm': 1000000, 'g': 1000000000 }
```

```python
#  TODAS AS FUNÇÕES 
import re

def normalize_text(text):
    text = text.lower().replace(',','.').replace('\n','')
    text = re.sub('[ãáäà]','a',text)
    text = re.sub('[éẽèê]','e',text)
    text = re.sub('[ïíì]','i',text)
    text = re.sub('[õòó]','o',text)
    text = re.sub('[ùúü]','u',text)
    return text


#  FUNÇÃO QUE CONVERTE AS LETRAS EM NÚMEROS 

def convert_number_to_text(txt):
    txt = re.sub('[g]', ' g', txt)
    txt = re.sub('[k]', ' k', txt)
    txt = re.sub('[m]', ' m', txt)
    number = float(re.search('\d*.\d*', txt).group(0))
    

    num = re.search('[a-z]', txt).group(0)
    num = dict_str[num]
    return str(number*num)



def convert_text_to_number(txt):
    
    number_final = 0
    sum = 0
    number = 0
    txt = txt.split(' ')
    for word in txt:
        if word in dict_numbers.keys():
            sum += dict_numbers[word]
            number += sum

        elif word in dict_unidades.keys():
            if word == 'mil' and number == 0:
                number = 1
            number *= dict_unidades[word]
            number_final += number
            number = 0
        sum = 0
    number_final += number
    return str(number_final)

# Função para ler o arquivo

def funcao_final():
    arquivo = open('input.txt', 'r', encoding = 'utf-8')
    output = open('output.txt', 'w', encoding = 'utf-8')
    
    for linha in arquivo:
        if re.match('\d',linha):
            output.writelines(convert_number_to_text(normalize_text(linha))+'\n')
        else:
            output.writelines(convert_text_to_number(normalize_text(linha))+'\n')
            
#FINAL DO CODIGO CHAMANDO O COMANDO
funcao_final()
```
